```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
from scipy import stats
```

```python
#Abrir o arquivo
df = pd.read_csv('LoL.csv')
#listar colunas
df.head()
```

```python
bwin = df["bResult"].value_counts()[0]
rwin = df["bResult"].value_counts()[1]
print("Vitórias time azul: {}" .format(bwin))
print("Vitórias time vermelho: {}" .format(rwin))
```

```python
df["redTeamTag"].str.upper()
```

```python
print("Times e quantidades de jogos:")
print(df["redTeamTag"].value_counts())
```

```python
dic = {}
b = sorted(df["redTeamTag"].unique())

for i in b:
    if i not in dic:
        dic[i] = [0,0]
    for j in range(len(df)):
        if df["redTeamTag"][j] == i:
            if df["rResult"][j] == 1:
                dic[i][0] += 1
            else:
                dic[i][1] += 1
```

```python
print("Times e as quantidades de vitórias e derrotas:")
for i in dic:
    print(i, dic[i])
```
